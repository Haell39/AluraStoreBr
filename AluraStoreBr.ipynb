### Importação dos dados



```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

#1. Análise do faturamento


```python
# Calcular o faturamento total para cada loja
faturamento_loja_1 = loja['Preço'].sum()
faturamento_loja_2 = loja2['Preço'].sum()
faturamento_loja_3 = loja3['Preço'].sum()
faturamento_loja_4 = loja4['Preço'].sum()

# DataFrame para exibir o faturamento total de cada loja
faturamento_total_df = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento Total': [faturamento_loja_1, faturamento_loja_2, faturamento_loja_3, faturamento_loja_4]
})

print("Análise do Faturamento Total das Lojas:\n")
print(faturamento_total_df)
```

```python
# Gráfico de Pizza para Proporção do Faturamento
plt.figure(figsize=(8, 8))
plt.pie(faturamento_total_df['Faturamento Total'], labels=faturamento_total_df['Loja'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proporção do Faturamento Total por Loja')
plt.axis('equal')
plt.tight_layout()
plt.savefig('faturamento_total_por_loja_pizza.png')
print("Gráfico de Pizza 'faturamento_total_por_loja_pizza.png' gerado.")
```

# 2. Vendas por Categoria


```python
# Concatenar todos os DataFrames das lojas em um único DataFrame
all_stores_df = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Calcular vendas por categoria
vendas_por_categoria = all_stores_df.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

print("\nAnálise de Vendas por Categoria (Todas as Lojas):\n")
print(vendas_por_categoria)
```

```python
# Gerar gráfico de vendas por categoria
plt.figure(figsize=(12, 7))
sns.barplot(x=vendas_por_categoria.index, y=vendas_por_categoria.values, palette='magma')
plt.title('Vendas por Categoria de Produto (Todas as Lojas)')
plt.xlabel('Categoria do Produto')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('vendas_por_categoria.png')
print("Gráfico 'vendas_por_categoria.png' gerado.")
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média de avaliação para cada loja
media_avaliacao_loja_1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja_2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja_3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja_4 = loja4['Avaliação da compra'].mean()

# criando o DataFrame para exibir a média de avaliação de cada loja
media_avaliacao_df = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média de Avaliação': [media_avaliacao_loja_1, media_avaliacao_loja_2, media_avaliacao_loja_3, media_avaliacao_loja_4]
})

print("\nAnálise da Média de Avaliação das Lojas:\n")
print(media_avaliacao_df)
```

```python
# Gerar Dot Plot para média de avaliação
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Média de Avaliação', y='Loja', data=media_avaliacao_df, s=200, color='blue', zorder=2)
for i, row in media_avaliacao_df.iterrows():
    plt.text(row['Média de Avaliação'] + 0.02, row['Loja'], f"{row['Média de Avaliação']:.2f}", va='center')
plt.axvline(x=media_avaliacao_df['Média de Avaliação'].mean(), color='gray', linestyle='--', label='Média Geral')
plt.title('Média de Avaliação das Compras por Loja')
plt.xlabel('Média de Avaliação (1-5)')
plt.ylabel('Loja')
plt.xlim(media_avaliacao_df['Média de Avaliação'].min() - 0.1, media_avaliacao_df['Média de Avaliação'].max() + 0.1)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig('media_avaliacao_lojas_dot_plot.png')
print("Gráfico Dot Plot")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Calcular os produtos mais vendidos
produtos_mais_vendidos = all_stores_df['Produto'].value_counts().head(10)

# Calcular os produtos menos vendidos
produtos_menos_vendidos = all_stores_df['Produto'].value_counts().tail(10)

print("\nAnálise dos Produtos Mais Vendidos (Todas as Lojas): \n")
print(produtos_mais_vendidos)

print("\nAnálise dos Produtos Menos Vendidos (Todas as Lojas):\n")
print(produtos_menos_vendidos)
```

```python
# Gráfico para Produtos Mais Vendidos
plt.figure(figsize=(8, 6))
sns.barplot(x=produtos_mais_vendidos.index, y=produtos_mais_vendidos.values, palette='Greens_d')
plt.title('Top 10 Produtos Mais Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('top_10_produtos_mais_vendidos.png')
print("Gráfico de Barras 'top_10_produtos_mais_vendidos.png' gerado.")
```

```python
# Gráfico para Produtos Menos Vendidos
plt.figure(figsize=(8, 6))
sns.barplot(x=produtos_menos_vendidos.index, y=produtos_menos_vendidos.values, palette='Reds_d')
plt.title('Top 10 Produtos Menos Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('top_10_produtos_menos_vendidos.png')
print("Gráfico de Barras 'top_10_produtos_menos_vendidos.png' gerado.")
```

# 5. Frete Médio por Loja


```python
# Calcular o frete médio para cada loja
frete_medio_loja_1 = loja['Frete'].mean()
frete_medio_loja_2 = loja2['Frete'].mean()
frete_medio_loja_3 = loja3['Frete'].mean()
frete_medio_loja_4 = loja4['Frete'].mean()

# Criar um DataFrame para exibir o frete médio de cada loja
frete_medio_df = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': [frete_medio_loja_1, frete_medio_loja_2, frete_medio_loja_3, frete_medio_loja_4]
})

print("\nAnálise do Frete Médio por Loja:\n")
print(frete_medio_df)
```

```python
# Gerar gráfico de frete médio ==> Barras Horizontais
plt.figure(figsize=(10, 6))
sns.barplot(x='Frete Médio', y='Loja', data=frete_medio_df, palette='plasma') # Eixos trocados
plt.title('Frete Médio por Loja')
plt.xlabel('Frete Médio (R$)') # Rótulo do eixo X
plt.ylabel('Loja') # Rótulo do eixo Y
plt.grid(axis='x', linestyle='--', alpha=0.7) # Grid no eixo X
plt.tight_layout()
plt.savefig('frete_medio_por_loja_horizontal.png') # Novo nome para diferenciar
print("Gráfico 'frete_medio_por_loja_horizontal.png' gerado.")
```

# Relatório de Análise de Desempenho das Lojas do Seu João

## Introdução

Este relatório apresenta uma análise do desempenho financeiro e operacional das quatro lojas do Seu João, com o objetivo de fornecer insights para uma decisão estratégica sobre qual loja, se houver, deve ser considerada para venda. A análise baseou-se em dados de faturamento, categorias de produtos, avaliações de clientes, volume de vendas de produtos específicos e custos de frete.

## Dados Utilizados

Foram utilizados quatro conjuntos de dados em formato CSV, um para cada loja (`loja.csv`, `loja2.csv`, `loja3.csv`, `loja4.csv`), contendo detalhes transacionais de vendas.

## Resultados das Análises

### Análise do Faturamento Total

- **Loja 1**: R$1.534.509,12 (**Maior Faturamento**)
- **Loja 2**: R$1.488.459,06
- **Loja 3**: R$1.464.025,03
- **Loja 4**: R$1.384.497,58 (**Menor Faturamento**)

A Loja 1 demonstrou o maior volume de vendas em termos de receita, enquanto a Loja 4 apresentou o menor faturamento.

### Vendas por Categoria (Todas as Lojas Combinadas)

As categorias de **eletrônicos** (aproximadamente R$2,21 milhões) e **eletrodomésticos** (aproximadamente R$1,76 milhões) são as que mais contribuem para o faturamento geral das lojas. As categorias de menor venda foram _livros_ e _utilidades domésticas_.

### Média de Avaliação das Lojas

- **Loja 3**: 4,05 (**Maior Média de Avaliação**)
- **Loja 2**: 4,04
- **Loja 4**: 4,00
- **Loja 1**: 3,98 (**Menor Média de Avaliação**)

A Loja 3 se destaca pela maior satisfação do cliente, enquanto a Loja 1 apresenta a menor média de avaliação, sugerindo possíveis áreas de melhoria no atendimento ou qualidade.

### Produtos Mais e Menos Vendidos (Todas as Lojas Combinadas)

- **Mais Vendidos (Top 3)**: Cômoda, Micro-ondas e Carrinho controle remoto.
- **Menos Vendidos (Bottom 3)**: Celular ABXY, Headset e Mochila.

Esses dados indicam os produtos de maior e menor demanda no conjunto das lojas.

### Frete Médio por Loja

- **Loja 1**: R$34,69 (**Maior Frete Médio**)
- **Loja 2**: R$33,62
- **Loja 3**: R$33,07
- **Loja 4**: R$31,28 (**Menor Frete Médio**)

A Loja 4 demonstra a maior eficiência em custos de frete, enquanto a Loja 1 tem o custo médio de frete mais alto.

## Recomendação Final

Com base nas análises realizadas e na descrição da atividade, a recomendação para Seu João, caso o objetivo principal seja vender a loja com o menor desempenho financeiro direto e otimizar o portfólio, é considerar a **Loja 4** para venda.

### Justificativa:

- A Loja 4 consistentemente apresentou o menor faturamento total entre as quatro lojas. Embora demonstre eficiência no frete (menor frete médio) e tenha uma avaliação de cliente razoável, sua contribuição para a receita geral é a mais baixa.
- As Lojas 2 e 3 demonstram um bom equilíbrio entre faturamento e alta satisfação do cliente (Loja 3 sendo a líder em avaliações), além de custos de frete competitivos. Elas representam ativos sólidos no portfólio.
- A Loja 1, apesar de ter o maior faturamento, também apresenta a menor avaliação dos clientes e o maior frete médio. Isso sugere que, embora gere alta receita, pode haver desafios operacionais ou de satisfação que exigiriam investimento para serem superados. No entanto, o alto volume de vendas a torna um pilar importante da receita total.

**Importante**: Se a decisão de venda for puramente baseada em **eliminar o ativo de menor receita**, a *Loja 4* é a candidata mais direta, porem, se o foco for em manter a maior parte da receita e ainda assim, focar nas lojas com um **exelente potencial para crescer** e gerar maior receita no futuro, a *loja 1* deve ser considerada como opção principal para venda pela sua ineficiencia com o frete e menor satisfação dos clientes.
